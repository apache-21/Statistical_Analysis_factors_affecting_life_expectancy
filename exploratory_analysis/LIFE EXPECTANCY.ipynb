#### Relationship between features and life expectancy

### _Should a country having a lower life expectancy value (<65 years) increase its healthcare expenditure in order to improve its average lifespan?_


```python
df['Health expenditure'] = (df['percentage expenditure']/100)*df['GDP']
df.head()
```

```python
print(df['Life expectancy'].min(), df['Life expectancy'].max())
```

```python

df['Expectancy bin'] = pd.cut(df['Life expectancy'], bins=(35,65,90), labels=('less than 65', 'greater than 65'))
df['Expectancy bin'] = df['Expectancy bin'].astype('str')
df.groupby('Expectancy bin')['Health expenditure'].mean().plot.bar()
```

Countries with life expectancy under 65 have a very low health expenditure compared to countries with over 65yrs  life expectancy and thus needs to increase their health expenditure

```python
(df.groupby(['Expectancy bin', 'Status'])['Country'].count()/16).plot.bar()
# dividing by 16 since the dataset is spread across 16yrs to get a correct count of countries
```

There is no developed country with life expectancy less than 65 years 

### Analysis on country status to see the difference in features

### Chloropeth map visualization of life expectancy

```python
import folium
World_map = folium.Map(location=(14.5994, 28.6731), 
                       zoom_start=1.3, zoom_control=True,
                       scrollWheelZoom=False,
                       dragging=True, min_zoom=1,max_zoom=3,)
```

```python
World_map.choropleth(
    geo_data=r"C:\Users\Atehe Stephen\Documents\world.json",
    data=df,
    columns=['Country','Life expectancy'],
    key_on='feature.properties.name',
    fill_color='YlGnBu',
    fill_opacity=1,
    line_opacity=0.2,
    legend_name="Life expectancy",
    smooth_factor=0,
    Highlight= True,
    line_color = "Black",
    nan_fill_color = "black")
```
