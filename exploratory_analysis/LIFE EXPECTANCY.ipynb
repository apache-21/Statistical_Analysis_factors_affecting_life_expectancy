```python
#Remove whitespace in the columns name
for col in df.columns:
    df.rename(columns={col:col.strip()}, inplace=True)
```

```python
sns.set(rc={'figure.figsize':(12,9)})
sns.heatmap(df.corr())
```

```python
# scatterplot of life expectancy and the features 
for col in df.select_dtypes(include='number').columns:
    fig = plt.figure(figsize=(6,6))
    ax = fig.gca()
    sns.regplot(x=col, y='Life expectancy', data=df)
    plt.title(col.upper())
    plt.show()
   
```

```python
stat_df = df.groupby('Status').mean()
```

```python
stat_df['Life expectancy'].plot.bar()
```

Comparing of features between the developeed and developing countries

```python
stat_df_ = stat_df[stat_df < 200]
stat_df_.dropna(inplace=True, axis=1)
stat_df_.plot.bar()
```

Alcohol doesnt seem to affect life expectancy negatively.

```python
#plot for other features 
for col in df.columns:
    if col not in stat_df_.columns and df[col].dtypes != 'O' :
        fig = plt.figure(figsize=(6,6))
        ax = fig.gca()
        stat_df[col].plot.bar()
        plt.title(col.upper())
        plt.show()
        
        
```
