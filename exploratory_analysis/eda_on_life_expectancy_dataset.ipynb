```python
# Calculate correlations between numeric features
correlations = df.corr()
```

```python
# Increase the figsize to 7 x 6
plt.figure(figsize=(9,9))

# Plot heatmap of correlations
sns.heatmap(correlations, cmap='RdBu_r')
plt.show()
```

The above heatmap could not tells us because the color seems to be much to pick the correlations appears difficulty. Hence we will use a triangular heatmap with percentage to see the correlations among the numeric features

```python
# Generate a mask for the upper triangle
mask = np.zeros_like(correlations)
mask[np.triu_indices_from(mask)] = 1

# Option 2: Mask of False/True
# mask = np.zeros_like(correlations, dtype=np.bool)
# mask[np.triu_indices_from(mask)] = True
```

```python
# Make the figsize 10 x 8
plt.figure(figsize=(10,8))

# Plot heatmap of correlations
sns.heatmap(correlations * 100,
            cmap='RdBu_r',
            annot=True,
            fmt='.0f',
            mask=mask)
plt.show()
```

Insight from the above plot:
The following appears to be highly correlated with target variable-
* schooling
* income_composition_of_resources
* adult_mortality ( -ve correlation)


The following features appears to be correlated with each-
* infant_deaths and under_five_deaths are strongly correlated
* percentage_expenditure and gdp
*  some correlation between hepatitis b and diphtheria
* income_composition_resources and schooling
* thinness 1_19 and thinness 5_9 
* polio and diphtheria seems to be correlated fairly

```python
# Examining the relationship between the status and life_expectancy

sns.violinplot(y='status', x='life_expectancy', data=df)
plt.show()
```

```python
# Scatterplot of schooling vs. .life expectancy
sns.lmplot(x='schooling',
           y='life_expectancy',
           hue='status',
           data=df,
           fit_reg=False,
           scatter_kws={'alpha':0.1})
plt.show()
```

An upward trends is seen- this imply as the schooling rate increase the life expectancy increase, moreso developed country have high life_expectancy and the schoolling rate is high

```python
# The relationship between status and income_composition_of_resources
sns.boxplot(y='status', x='income_composition_of_resources', data=df)
plt.show()
```

The income compostion of resources is high in developed countries and their appears to be outlier in the Developing countries and this is below the minimun value . could the dataset contain negative values for income?

```python
# The income distribution of the income 
sns.violinplot(y='status', x='income_composition_of_resources', data=df)
plt.show()
```

The above shows how the income of developed and developing countries are distributed
* for developing country : there appears to income division class - from 0.0 -0.1, 0.3 -0.6 and 0.6 - 0.9

```python
# Scatterplot of income_composition vs expectancy
sns.lmplot(x='income_composition_of_resources',
           y='life_expectancy',
           hue='status',
           data=df,
           fit_reg=False,
           scatter_kws={'alpha':0.1})
plt.show()
```

```python
# relationship between  adult mortality and life expectancy for developed and developing countries
sns.lmplot(x='adult_mortality',
           y='life_expectancy',
           hue='status',
           data=df,
           fit_reg=False,
           scatter_kws={'alpha':0.1})
plt.show()
```
